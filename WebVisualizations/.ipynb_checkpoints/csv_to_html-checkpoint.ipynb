```python
# Import dependencies
import pandas as pd
```

```python
# Load file
file_path = "Resources/Weather_Data.csv"
```

```python
# Store csv file data in dataframe
weather_data_df = pd.read_csv(file_path)

# Print dataframe
weather_data_df
```

```python
# Set index to be the city ID
weather_data_df.set_index(["City_ID"], inplace = True)

# Print dataframe with new index
weather_data_df
```

```python
weather_data_df = weather_data_df.reset_index()
```

```python
# Convert dataframe to html
weather_data_html = weather_data_df.to_html("weather_data.html", index = False)

# Print html data
weather_data_html
```
